Summary of challenge: 
The Day 1 input contains a series of numbers which represent the number of calories each elf is carrying. Each line of text correspond to one elf. In part 1, we need to calculate the highest number of calories that an elf is carrying. In part 2, we need to calculate the sum of the top 3 highest numbers of calories that elves are carrying.

```python
import pandas as pd
```

Part 1

What is the highest number of calories that an elf is carrying?

```python
# open Day1_input as file and store the lines of the file as a string

strings = ''
with open("Day1_input.txt") as file:
  for line in file:
    if line is not None:
      strings+=line
strings
```

```python
# Create a list that splits the strings into seperate elements based on row separation ('\n\n')

lists = strings.split('\n\n')
lists
```

```python
# Turn the lists into a dataframe (df) that contains each element in a seperate column

df = pd.DataFrame(lists)
df=df.T # .T transposes the dataframe
df
```

```python
# Remove the '\n' from each column to generate seperate the numbers 
# note_1 - the numbers are still stored as strings
# note_2 - the dataframe is now a list

df = df.iloc[0].str.split('\n')
df
```

```python
# Changing the df list back to a dataframe
# The df now looks much tidier, although the numbers are still stored as strings

df = pd.DataFrame(df)
df=df.T
df.head()
```

```python
# The code does the following:
# (1) Converts the numbers (stored as strings) from the dataframe into floats
# (2) Stores the numbers into a new dataframe (floats) row by row (instead of one row containing multiple numbers)

floats=pd.DataFrame()
for i in range(0,239):
  # creating a new list of elements called elements
  elements=[]
  for x in df[i][0]:
    try:
      # adding the numbers to the elements and converting the numbers into floats
      elements.append(float(x))
    except:
      # ValueError might occur due to rogue empty ('') string
      ValueError
      continue
  try:
    # adding the elements into the appropriate column row by row
    floats[i]=elements
  except:
    # ValueError might occur due to differences in row numbers between different columns
    ValueError
    # By-passing the differences in row numbers by creating a new dataframe (test) which is empty until the exception is invoked
    test=pd.DataFrame()
    # adding the elements into the appropriate column row by row
    test[i]=elements
    # concatentating the two dataframes
    floats = pd.concat([floats, test], ignore_index=True)
```

```python
# The number of columns is correct (239)

floats.shape
```

```python
# There are a lot of nulls in the new dataframe due to the mismatch in the number of rows between different columns
# This is irrelevant in this scenario because all the neccessary data is in the dataframe, just in various indexes

floats.isnull().sum()
```

```python
# The highest number of calories that an elf is carrying is 69177.

max(floats.sum())
```

Part 2

What is the sum of the top 3 highest numbers of calories that elves are carrying?

```python
# creating a list of sums

sums=[]
for i in floats.sum():
  sums.append(i)
```

```python
# determining the position of max

for i in range(0,239):
  if sums[i]==69177.0:
    print(i)
```
